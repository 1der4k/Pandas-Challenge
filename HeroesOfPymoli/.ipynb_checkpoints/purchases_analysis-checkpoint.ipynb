```python
# Dependencies
import pandas as pd

# Load file
data_file = "../Resources/purchase_data.csv"

# Read file
purchasing_data =pd.read_csv(data_file)
```

```python
# Create dataframe
purchasing_data_df = pd.DataFrame(purchasing_data)

# Verify
purchasing_data_df.head(10)
```

```python
# Total number of players
# Find unique values in SN column of base dataframe
unique_users = purchasing_data_df["SN"].unique()
# Assign list of unique values to new dataframe in order to find count
unique_users_df = pd.DataFrame(unique_users)
# Find count of values in new dataframe
total_users=unique_users_df.count()
# Assign to new dataframe in order to label column as "Total Users"
total_users_df = pd.DataFrame({"Total Users":total_users})

total_users_df

```

```python
# Run basic calculations to obtain number of unique items, average purchase price, total number of purchases, total revenue
# Find total number of unique values in Item name column of base dataframe
unique_items = purchasing_data_df["Item Name"].unique()
unique_items_df = pd.DataFrame(unique_items)
unique_items_df
```

```python
total_unique_items = unique_items_df[0].count()
total_unique_items
#total_unique_items_df = pd.DataFrame([total_unique_items])
#total_unique_items_df
```

```python
# Find average of values in Price column of base dataframe
average_price = purchasing_data_df["Price"].mean()
#average_price_df = pd.DataFrame([average_price])
#average_price_df
```

```python
# Find total number of purchases
total_purchases = purchasing_data_df["Purchase ID"].count()
#total_purchases_df = pd.DataFrame([total_purchases])
#total_purchases_df
```

```python
# Find total revenue
total_revenue = purchasing_data_df["Price"].sum()
#total_revenue_df = pd.DataFrame([total_revenue])
#total_revenue_df
```

```python
# Create summary dataframe for display
purchasing_summary_df = pd.DataFrame({
    "Number of Unique Items":[total_unique_items],
    "Average Price":[average_price],
    "Number of Purchases":[total_purchases],
    "Total Revenue":[total_revenue]
                                    })
purchasing_summary_df
```

```python
format_purchasing_summary = purchasing_summary_df.copy()

format_purchasing_summary["Average Price"] = format_purchasing_summary["Average Price"].map("${:,.2f}".format)
format_purchasing_summary["Total Revenue"] = format_purchasing_summary["Total Revenue"].map("${:,.2f}".format)

format_purchasing_summary
```

```python
# Gender Demographics
sn_and_gender = purchasing_data_df[["SN", "Gender"]]
unique_sn_and_gender = sn_and_gender.drop_duplicates(subset="SN", keep="first")
users_groupby_gender = unique_sn_and_gender.groupby(["Gender"])
users_groupby_gender.count()
users_groupby_gender_df = pd.DataFrame(users_groupby_gender.count())
users_groupby_gender_df = users_groupby_gender_df.rename(columns={"SN": "Total Count"})
percentage_by_gender = users_groupby_gender_df["Total Count"]/users_groupby_gender_df["Total Count"].sum() * 100
users_groupby_gender_df["Percentage of Players"] = percentage_by_gender
users_groupby_gender_df["Percentage of Players"] = users_groupby_gender_df["Percentage of Players"].map("{:.2f}%".format)
users_groupby_gender_df
```

```python
# Repeated displapy of base dataframe for convenience
purchasing_data_df.head(10)
```

```python
# Purchasing Analysis by Gender
# Extract useful columns from base dataframe
purchasing_user_gender_df = purchasing_data_df[["Gender","Purchase ID","Price","SN"]]
purchasing_user_gender_df
```

```python
# Group columns by gender
purchasing_groupby_gender = purchasing_user_gender_df.groupby(["Gender"])
purchasing_groupby_gender.head()
```

```python
# Count of columns gives count of purchases
purchasing_groupby_gender_df = purchasing_groupby_gender.count()
purchasing_groupby_gender_df = purchasing_groupby_gender_df.rename(columns={"Purchase ID":"Purchase Count"})
purchasing_groupby_gender_df
```

```python
# Find mean of price of items purchases by gender
average_price_by_gender_df = pd.DataFrame(purchasing_groupby_gender["Price"].mean())
purchasing_groupby_gender_df["Average Purchase Price"] = average_price_by_gender_df
purchasing_groupby_gender_df["Average Purchase Price"] = purchasing_groupby_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_groupby_gender_df
```

```python
# Find sum of price of purchases by gender
total_price_gender = purchasing_groupby_gender["Price"].sum()
purchasing_groupby_gender_df["Total Purchase Value"] = total_price_gender
purchasing_groupby_gender_df["Total Purchase Value"]= purchasing_groupby_gender_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_groupby_gender_df
```

```python
# Mean purchase total per person
# Extract relevant columsn from base dataframe, then use loc to separate rows by gender and determine mean total user purchase

price_by_gender_sn_df = purchasing_data_df[["Gender", "SN","Price"]]

# Female
price_by_female_sn_df = price_by_gender_sn_df.loc[price_by_gender_sn_df["Gender"]=="Female", :]
price_groupby_female_sn = price_by_female_sn_df.groupby("SN")
sum_price_groupby_female_sn_df = pd.DataFrame(price_groupby_female_sn.sum())
female_user_purchase_mean_df = sum_price_groupby_female_sn_df.mean()
female_user_purchase_mean_df
```

```python
# Male
price_by_male_sn_df = price_by_gender_sn_df.loc[price_by_gender_sn_df["Gender"]=="Male", :]
price_groupby_male_sn = price_by_male_sn_df.groupby("SN")
sum_price_groupby_male_sn_df = pd.DataFrame(price_groupby_male_sn.sum())
male_user_purchase_mean_df = sum_price_groupby_male_sn_df.mean()
male_user_purchase_mean_df
```

```python
# Other
price_by_other_sn_df = price_by_gender_sn_df.loc[price_by_gender_sn_df["Gender"]=="Other / Non-Disclosed", :]
price_groupby_other_sn = price_by_other_sn_df.groupby("SN")
sum_price_groupby_other_sn_df = pd.DataFrame(price_groupby_other_sn.sum())
other_user_purchase_mean_df = sum_price_groupby_other_sn_df.mean()
other_user_purchase_mean_df
```

```python
# Create series of mean total user purchase by gender
mean_total_purchase_per_user = pd.Series({"Avg Total Purchase per Person":[female_user_purchase_mean_df[0],male_user_purchase_mean_df[0],other_user_purchase_mean_df[0]]})
mean_total_purchase_per_user
```

```python
purchasing_groupby_gender_df["Avg Total Purchase per Person"] = mean_total_purchase_per_user["Avg Total Purchase per Person"]
purchasing_groupby_gender_df
```

```python
# Purchase analysis by gender summary formatting and display
purchasing_groupby_gender_df = purchasing_groupby_gender_df.drop(columns=["Price","SN"])
purchasing_groupby_gender_df["Avg Total Purchase per Person"] = purchasing_groupby_gender_df["Avg Total Purchase per Person"].map("${:.3}".format)
purchasing_groupby_gender_df
```

```python
# Age Demographics
# Bins
bins = [0, 9.9, 13.9, 19.9, 24.9, 29.9, 34.9, 39.9, 49.9]
labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
user_age_df = purchasing_data_df[["SN","Age"]]
unique_user_age_df = user_age_df.drop_duplicates(subset="SN", keep="first")
unique_user_age_df["Age"] = pd.cut(unique_user_age_df["Age"],bins, labels = labels, include_lowest=True )
unique_user_groupby_age = unique_user_age_df.groupby("Age")
unique_user_groupby_age_df = pd.DataFrame(unique_user_groupby_age.count())
unique_user_groupby_age_df
```

```python
percentage_by_age= unique_user_groupby_age_df["SN"]/unique_user_groupby_age_df["SN"].sum() * 100
unique_user_groupby_age_df["Percentage of Players"] = percentage_by_age
unique_user_groupby_age_df = unique_user_groupby_age_df.rename(columns={"SN":"Total Count"})
unique_user_groupby_age_df["Percentage of Players"] = unique_user_groupby_age_df["Percentage of Players"].map("{:.2f}%".format)
unique_user_groupby_age_df
```

```python
# Purchasing Analysis by Age
bins = [0, 9.9, 13.9, 19.9, 24.9, 29.9, 34.9, 39.9, 49.9]
labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchasing_data_age_df = purchasing_data_df[["Purchase ID","Age","Price"]]
purchasing_data_age_df["Age"] = pd.cut(purchasing_data_age_df["Age"],bins, labels = labels, include_lowest=True)
purchases_groupby_age = purchasing_data_age_df.groupby("Age")
purchases_groupby_age.head()
```

```python
purchases_groupby_age_df = pd.DataFrame(purchases_groupby_age.count())
purchases_groupby_age_df = purchases_groupby_age_df.rename(columns={"Purchase ID": "Purchase Count"})
purchases_groupby_age_df
```

```python
mean_age_purchase = purchases_groupby_age["Price"].mean()
purchases_groupby_age_df["Average Purchase Price"] = mean_age_purchase
purchases_groupby_age_df
total_age_purchase = purchases_groupby_age["Price"].sum()
purchases_groupby_age_df["Total Purchase Value"] = total_age_purchase
purchases_groupby_age_df
```

```python
# avg total purchase per person

price_by_age_sn_df = purchasing_data_df[["SN","Age","Price"]]

# Under 10
price_by_underten_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 10, :]
price_by_underten_sn_df
price_groupby_underten_sn = price_by_underten_sn_df.groupby("SN")
sum_price_groupby_underten_sn_df = pd.DataFrame(price_groupby_underten_sn["Price"].sum())
underten_user_purchase_mean_df = sum_price_groupby_underten_sn_df.mean()
underten_user_purchase_mean_df
```

```python
# under 14

price_by_underfourteen_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 14, :]
price_by_underfourteen_sn_df
price_groupby_underfourteen_sn = price_by_underfourteen_sn_df.groupby("SN")
sum_price_groupby_underfourteen_sn_df = pd.DataFrame(price_groupby_underfourteen_sn["Price"].sum())
underfourteen_user_purchase_mean_df = sum_price_groupby_underfourteen_sn_df.mean()

# under 19

price_by_undernineteen_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 19, :]
price_by_undernineteen_sn_df
price_groupby_undernineteen_sn = price_by_undernineteen_sn_df.groupby("SN")
sum_price_groupby_undernineteen_sn_df = pd.DataFrame(price_groupby_undernineteen_sn["Price"].sum())
undernineteen_user_purchase_mean_df = sum_price_groupby_undernineteen_sn_df.mean()

# under 24

price_by_undertwentyfour_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 24, :]
price_by_undertwentyfour_sn_df
price_groupby_undertwentyfour_sn = price_by_undertwentyfour_sn_df.groupby("SN")
sum_price_groupby_undertwentyfour_sn_df = pd.DataFrame(price_groupby_undertwentyfour_sn["Price"].sum())
undertwentyfour_user_purchase_mean_df = sum_price_groupby_undertwentyfour_sn_df.mean()

# under 29

price_by_undertwentynine_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 29, :]
price_by_undertwentynine_sn_df
price_groupby_undertwentynine_sn = price_by_undertwentynine_sn_df.groupby("SN")
sum_price_groupby_undertwentynine_sn_df = pd.DataFrame(price_groupby_undertwentynine_sn["Price"].sum())
undertwentynine_user_purchase_mean_df = sum_price_groupby_undertwentynine_sn_df.mean()

# under 34

price_by_underthirtyfour_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 34, :]
price_by_underthirtyfour_sn_df
price_groupby_underthirtyfour_sn = price_by_underthirtyfour_sn_df.groupby("SN")
sum_price_groupby_underthirtyfour_sn_df = pd.DataFrame(price_groupby_underthirtyfour_sn["Price"].sum())
underthirtyfour_user_purchase_mean_df = sum_price_groupby_underthirtyfour_sn_df.mean()

# under 39

price_by_underthirtynine_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] <= 39, :]
price_by_underthirtynine_sn_df
price_groupby_underthirtynine_sn = price_by_underthirtynine_sn_df.groupby("SN")
sum_price_groupby_underthirtynine_sn_df = pd.DataFrame(price_groupby_underthirtynine_sn["Price"].sum())
underthirtynine_user_purchase_mean_df = sum_price_groupby_underthirtynine_sn_df.mean()

# over 40
price_by_overforty_sn_df = price_by_age_sn_df.loc[price_by_age_sn_df["Age"] >= 40, :]
price_by_overforty_sn_df
price_groupby_overforty_sn = price_by_overforty_sn_df.groupby("SN")
sum_price_groupby_overforty_sn_df = pd.DataFrame(price_groupby_overforty_sn["Price"].sum())
overforty_user_purchase_mean_df = sum_price_groupby_overforty_sn_df.mean()
overforty_user_purchase_mean_df
```

```python
# Create series of mean total user purchase by age
mean_total_purchase_age_user = pd.Series({"Avg Total Purchase per Person":[
    underten_user_purchase_mean_df[0],
    underfourteen_user_purchase_mean_df[0],
    undernineteen_user_purchase_mean_df[0],
    undertwentyfour_user_purchase_mean_df[0],
    undertwentynine_user_purchase_mean_df[0],
    underthirtyfour_user_purchase_mean_df[0],
    underthirtynine_user_purchase_mean_df[0],
    overforty_user_purchase_mean_df[0]
                    ]})
mean_total_purchase_age_user
```

```python
# FINAL PURCHASE ANALYSIS BY AGE SUMMARY
purchases_groupby_age_df["Avg Total Purchase per Person"] = mean_total_purchase_age_user["Avg Total Purchase per Person"]
purchases_groupby_age_df = purchases_groupby_age_df.drop(columns=["Price"])
purchases_groupby_age_df["Average Purchase Price"] = purchases_groupby_age_df["Average Purchase Price"].map("${:,.2f}".format)
purchases_groupby_age_df["Total Purchase Value"] = purchases_groupby_age_df["Total Purchase Value"].map("${:,.2f}".format)
purchases_groupby_age_df["Avg Total Purchase per Person"] = purchases_groupby_age_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchases_groupby_age_df
```
